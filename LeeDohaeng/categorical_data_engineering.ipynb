# F-value test

```python
df_train = pd.read_csv("train_no_missing.csv")
df_test = pd.read_csv("test_no_missing.csv")
```

```python
nuTocaList = ["MSSubClass", "MoSold", "YrSold", "OverallQual"]
df_train[nuTocaList] = df_train[nuTocaList].astype("str")
df_test[nuTocaList] = df_test[nuTocaList].astype("str")
```

```python
df_all = pd.concat([df_train, df_test])
df_all.reset_index(drop = True, inplace = True)

df_all = pd.get_dummies(df_all)
df_train_dummy = df_all.iloc[:1460]
df_test_dummy = df_all.iloc[1460:]

df_test_dummy.drop("SalePriceLog", axis = 1, inplace = True)

#get numeric data and categoric data
df_train_dummy_categoric_list = list(df_train_dummy.filter(regex="_").columns)
df_train_dummy_numeric_list = list(set(df_train_dummy.columns) - set(df_train_dummy_categoric_list))
df_train_dummy_numeric_list.remove("SalePriceLog")
df_train_dummy_numeric_list.remove("Id")

print(len(df_train_dummy_categoric_list), len(df_train_dummy_numeric_list), len(list(df_train.columns)))
```

```python
df_train_numeric_list = []
df_train_categoric_list = []

for c in df_train.columns:
    if df_train[c].dtype == "int64" or df_train[c].dtype == "float64":
        df_train_numeric_list.append(c)
    else:
        df_train_categoric_list.append(c)

# df_train = pd.get_dummies(df_train)
# li_column = [c for c in df_train.columns]
        
df_train_numeric_list.remove("SalePriceLog")
df_train_numeric_list.remove("Id")

len(df_train_numeric_list)
```

```python
df_one_anova_category_Fvalue = pd.DataFrame(columns = ["name", "Fvalue"])

for idx, c in enumerate(df_train_categoric_list):
    model = sm.OLS.from_formula("SalePriceLog ~ C({})".format(c), data=df_train)
    result = model.fit()
    df_tmp=sm.stats.anova_lm(result)
    df_one_anova_category_Fvalue.loc[idx] = {"name":c, "Fvalue":df_tmp['PR(>F)'][0]}   

df_one_anova_category_Fvalue.sort_values(by=["Fvalue"], ascending=False, inplace=True)
df_one_anova_category_Fvalue.reset_index(drop=True, inplace=True)
df_one_anova_category_Fvalue
```

```python
df_one_anova_class_Fvalue = pd.DataFrame(columns = ["name", "Fvalue"])

for idx, c in enumerate(df_train_dummy_categoric_list):
    model = sm.OLS.from_formula("SalePriceLog ~ C({})".format(c), data=df_train_dummy)
    result = model.fit()
    df_tmp=sm.stats.anova_lm(result)
    df_one_anova_class_Fvalue.loc[idx] = {"name":c, "Fvalue":df_tmp['PR(>F)'][0]}   

df_one_anova_class_Fvalue.sort_values(by=["Fvalue"], ascending=False, inplace=True)
df_one_anova_class_Fvalue.reset_index(drop=True, inplace=True)
df_one_anova_class_Fvalue
```
